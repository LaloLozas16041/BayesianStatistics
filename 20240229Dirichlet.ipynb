### 01.- Ejemplo Uno

Aquí tienes un ejemplo en Python que muestra cómo trabajar con la distribución Dirichlet utilizando la librería scipy. Este ejemplo define una distribución Dirichlet tridimensional con parámetros $\alpha$ = (2,3,4) genera una muestra aleatoria y la visualiza en un gráfico 3D.

El código primero genera una muestra aleatoria de la distribución Dirichlet. Luego, crea una visualización 3D del simplex (que en tres dimensiones es un triángulo equilátero) donde se ubica el punto de la muestra. El punto rojo en el gráfico representa la muestra obtenida de la distribución Dirichlet.

La muestra obtenida es: $[0.19907408,0.21578935,0.58513657]$, lo que significa que estos son los valores que toman las variables aleatorias $x_1, x_2, x_3$ que suman 1, siguiendo la propiedad de la distribución Dirichlet.

```python
# Importar la librería necesaria para trabajar con la distribución Dirichlet
from scipy.stats import dirichlet
import numpy as np
import matplotlib.pyplot as plt

# Definir los parámetros alpha para la distribución Dirichlet
# En este ejemplo, se utilizan tres parámetros para una distribución tridimensional.
alpha_params = np.array([2, 3, 4])

# Crear una instancia de la distribución Dirichlet con los parámetros dados
dist = dirichlet(alpha_params)

# Generar una muestra aleatoria de la distribución
sample = dist.rvs(size=10)

# Visualizar la muestra en un espacio tridimensional
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Puntos para el simplex (triángulo equilátero en 3D)
simplex_points = np.array([[1, 0, 0], [0, 1, 0], [0, 0, 1]])

# Dibujar los puntos de la muestra en el simplex
ax.scatter(sample[:, 0], sample[:, 1], sample[:, 2], c='red', label='Muestra Dirichlet')

# Etiquetas y título
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$x_3$')
ax.set_title('Muestra de una Distribución Dirichlet')

# Leyenda
ax.legend()

# Mostrar gráfico
plt.show()

# Imprimir la muestra
print("Muestra de la distribución Dirichlet:", sample)
```

```python
# Generar una muestra más grande de la distribución Dirichlet para proporcionar una mejor ilustración.
# En este ejemplo, generaremos 500 puntos.

# Aumentar el tamaño de la muestra
sample_size = 500
samples = dist.rvs(size=sample_size)

# Visualizar las muestras en un espacio tridimensional
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Dibujar los puntos de la muestra en el simplex
ax.scatter(samples[:, 0], samples[:, 1], samples[:, 2], c='blue', alpha=0.5, label='Muestras Dirichlet')

# Etiquetas y título
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$x_3$')
ax.set_title('Muestras de una Distribución Dirichlet')

# Leyenda
ax.legend()

# Mostrar gráfico
plt.show()

# Imprimir las primeras 5 muestras para referencia
print("Primeras 5 muestras de la distribución Dirichlet:")
print(samples[:5])
```

### 03.- Ejemplo Tres

```python
# Ejemplo en Python para ilustrar la relación entre una distribución multinomial y Dirichlet

from scipy.stats import multinomial, dirichlet
import numpy as np

# Parámetros de la distribución multinomial
N = 100  # número total de ensayos
p = [0.3, 0.5, 0.2]  # probabilidades de éxito para cada uno de los k resultados

# Generar una muestra de la distribución multinomial
multinomial_sample = multinomial.rvs(N, p, size=1)

# Parámetros de la distribución Dirichlet (usualmente se basarían en conocimiento previo o datos anteriores)
alpha = [3, 5, 2]

# Crear una instancia de la distribución Dirichlet
dirichlet_dist = dirichlet(alpha)

# Actualizar los parámetros alpha con los resultados de la muestra multinomial
alpha_posterior = np.add(multinomial_sample, alpha)

# Generar una muestra de la distribución Dirichlet posterior
dirichlet_sample = dirichlet_dist.rvs(size=1)

# Imprimir resultados
print(f"Muestra de la distribución Multinomial (número de éxitos para cada resultado): {multinomial_sample}")
print(f"Muestra de la distribución Dirichlet (proporciones esperadas para cada resultado): {dirichlet_sample}")
print(f"Parámetros de la distribución Dirichlet posterior: {alpha_posterior}")
```

* Definimos una distribución multinomial con un total de $N=100$ ensayos y probabilidades de éxito para cada uno de los $k$ resultados dados por 
$p =[0.3,0.5,0.2]$

* Generamos una muestra de la distribución multinomial, que nos da el número de éxitos para cada uno de los resultados.

* Establecemos parámetros iniciales para la distribución Dirichlet $\alpha =[3,5,2]$, que podrían basarse en conocimiento o datos previos.

* Actualizamos estos parámetros con los resultados de la muestra multinomial para obtener la distribución Dirichlet posterior.

* Generamos una muestra de la distribución Dirichlet con los parámetros actualizados.

```python
# Vamos a ilustrar visualmente la relación entre las distribuciones multinomial y Dirichlet
# con un gráfico de barras para la muestra multinomial y un gráfico de la distribución Dirichlet.

# Parámetros iniciales
N = 100  # Total de ensayos
p = [0.3, 0.5, 0.2]  # Probabilidades de cada resultado
alpha = [3, 5, 2]  # Parámetros alpha de la distribución Dirichlet

# Crear la muestra multinomial
multinomial_sample = multinomial.rvs(N, p, size=1)[0]

# Crear la muestra Dirichlet a partir de los parámetros posteriores (alpha + observaciones)
alpha_posterior = np.add(multinomial_sample, alpha)
dirichlet_posterior_dist = dirichlet(alpha_posterior)
dirichlet_sample = dirichlet_posterior_dist.rvs(size=1)[0]

# Gráfico de barras para la muestra multinomial
plt.figure(figsize=(14, 5))

# Subplot para la muestra multinomial
plt.subplot(1, 2, 1)
plt.bar(range(1, len(multinomial_sample) + 1), multinomial_sample, color='blue', alpha=0.7)
plt.title('Muestra Multinomial')
plt.xlabel('Resultado')
plt.ylabel('Número de éxitos')
plt.xticks(range(1, len(multinomial_sample) + 1))

# Subplot para la muestra Dirichlet
plt.subplot(1, 2, 2)
plt.bar(range(1, len(dirichlet_sample) + 1), dirichlet_sample, color='orange', alpha=0.7)
plt.title('Proporciones de la Distribución Dirichlet Posterior')
plt.xlabel('Resultado')
plt.ylabel('Proporción')
plt.xticks(range(1, len(dirichlet_sample) + 1))

plt.tight_layout()
plt.show()

# Calcular y mostrar los insights
proporcion_observada = multinomial_sample / N
insights = {
    'Multinomial Sample Mean': np.mean(proporcion_observada),
    'Dirichlet Sample Mean': np.mean(dirichlet_sample),
    'Posterior Alpha': alpha_posterior,
    'Expected Proportion': alpha_posterior / np.sum(alpha_posterior)
}

for key, value in insights.items():
    print(f"{key}: {value}")
```


# Bayesian Odds Examples

## Introduction

This notebook explores the concept of odds from a Bayesian perspective, demonstrated through a series of examples. Odds in Bayesian analysis are used to update the probabilities of hypotheses based on new evidence. Each example will illustrate how Bayesian odds can be computed and interpreted in different scenarios.
            

```python

import numpy as np
import matplotlib.pyplot as plt

# Prior probability of hypothesis H being true
p_H = 0.6

# Probability of evidence E given H is true and H is false
p_E_given_H = 0.8
p_E_given_notH = 0.3

# Bayesian update using odds
odds_prior = p_H / (1 - p_H)
odds_posterior = (p_E_given_H / p_E_given_notH) * odds_prior
p_posterior = odds_posterior / (1 + odds_posterior)

# Printing updated probability
print(f"Posterior probability of H after observing E: {p_posterior:.2f}")

# Visualization
probabilities = [p_H, p_posterior]
positions = [0, 1]
labels = ['Prior', 'Posterior']
plt.bar(positions, probabilities, color=['blue', 'green'])
plt.xticks(positions, labels)
plt.ylabel('Probability')
plt.title('Bayesian Update of Beliefs')
plt.show()
        
```

This example demonstrates updating beliefs with Bayesian odds, which quantify the likelihood of a hypothesis relative to its alternative based on new evidence.

```python

import numpy as np

# Initial odds for two hypotheses H1 and H2
odds_H1 = 2.5
odds_H2 = 1.5

# Evidence favoring H1 over H2
likelihood_ratio = 1.8

# Updating odds with the likelihood ratio
updated_odds_H1 = odds_H1 * likelihood_ratio
updated_odds_H2 = odds_H2

# Calculating relative probability of H1 compared to H2
relative_probability_H1 = updated_odds_H1 / (updated_odds_H1 + updated_odds_H2)

print(f"Relative probability of H1 after evidence: {relative_probability_H1:.2f}")
        
```

This example illustrates using odds to compare and update multiple hypotheses based on the likelihood ratio from new evidence.

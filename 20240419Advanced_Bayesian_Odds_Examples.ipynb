
# Momios Bayesianos: Análisis Avanzado

## Introducción

Este notebook se enfoca en el uso de los momios desde una perspectiva Bayesiana, una herramienta fundamental en estadística para la actualización de probabilidades basada en nueva evidencia. En el contexto de la inferencia Bayesiana, los momios son una alternativa robusta al uso directo de probabilidades, proporcionando un enfoque más flexible para comparar hipótesis y actualizar creencias en respuesta a nuevos datos.

Cada ejemplo presentado aquí no solo demostrará cómo se calculan y actualizan los momios, sino que también explorará implicaciones más profundas y avanzadas de estos cálculos en la toma de decisiones y el análisis estadístico.


```python

import numpy as np
import matplotlib.pyplot as plt

# Prior probability of hypothesis H being true
p_H = 0.6

# Probability of evidence E given H is true and H is false
p_E_given_H = 0.8
p_E_given_notH = 0.3

# Bayesian update using odds
odds_prior = p_H / (1 - p_H)
odds_posterior = (p_E_given_H / p_E_given_notH) * odds_prior
p_posterior = odds_posterior / (1 + odds_posterior)

# Printing updated probability
print(f"Posterior probability of H after observing E: {p_posterior:.2f}")

# Visualization
probabilities = [p_H, p_posterior]
positions = [0, 1]
labels = ['Prior', 'Posterior']
plt.bar(positions, probabilities, color=['blue', 'green'])
plt.xticks(positions, labels)
plt.ylabel('Probability')
plt.title('Bayesian Update of Beliefs')
plt.show()
        
```

This example demonstrates updating beliefs with Bayesian odds, which quantify the likelihood of a hypothesis relative to its alternative based on new evidence.

```python

import numpy as np

# Initial odds for two hypotheses H1 and H2
odds_H1 = 2.5
odds_H2 = 1.5

# Evidence favoring H1 over H2
likelihood_ratio = 1.8

# Updating odds with the likelihood ratio
updated_odds_H1 = odds_H1 * likelihood_ratio
updated_odds_H2 = odds_H2

# Calculating relative probability of H1 compared to H2
relative_probability_H1 = updated_odds_H1 / (updated_odds_H1 + updated_odds_H2)

print(f"Relative probability of H1 after evidence: {relative_probability_H1:.2f}")
        
```

This example illustrates using odds to compare and update multiple hypotheses based on the likelihood ratio from new evidence.

```python

# Ejemplo 3: Evaluación de evidencia contradictoria
import numpy as np
import matplotlib.pyplot as plt

# Momios iniciales para una hipótesis H
odds_H = 3  # Implica una fuerte creencia inicial en H

# Evidencia a favor y en contra
p_E1_given_H = 0.9
p_E1_given_notH = 0.2
p_E2_given_H = 0.1
p_E2_given_notH = 0.8

# Actualización de momios con evidencia E1
odds_after_E1 = odds_H * (p_E1_given_H / p_E1_given_notH)

# Actualización de momios con evidencia E2
odds_after_E2 = odds_after_E1 * (p_E2_given_H / p_E2_given_notH)

# Probabilidad final después de toda la evidencia
final_probability = odds_after_E2 / (1 + odds_after_E2)

print(f"Probabilidad final de H: {final_probability:.2f}")

# Visualización de los cambios en los momios
odds_values = [odds_H, odds_after_E1, odds_after_E2]
labels = ['Inicial', 'Después de E1', 'Después de E2']
plt.plot(odds_values, marker='o')
plt.xticks(ticks=[0, 1, 2], labels=labels)
plt.title('Evolución de los Momios con Evidencia Contradictoria')
plt.ylabel('Momios')
plt.show()
        
```

Este tercer ejemplo analiza cómo los momios se actualizan con evidencias contradictorias, mostrando cómo una fuerte evidencia inicial a favor puede ser contrarrestada por evidencia posterior en contra.

```python

# Ejemplo 4: Comparación múltiple de hipótesis
import numpy as np

# Momios para tres hipótesis diferentes
odds_H1 = 1.5
odds_H2 = 2.0
odds_H3 = 0.5

# Evidencia que afecta diferencialmente a las hipótesis
likelihood_ratio_H1 = 1.2
likelihood_ratio_H2 = 0.8
likelihood_ratio_H3 = 2.0

# Actualización de momios
updated_odds_H1 = odds_H1 * likelihood_ratio_H1
updated_odds_H2 = odds_H2 * likelihood_ratio_H2
updated_odds_H3 = odds_H3 * likelihood_ratio_H3

# Cálculo de probabilidades relativas
total_odds = updated_odds_H1 + updated_odds_H2 + updated_odds_H3
p_H1 = updated_odds_H1 / total_odds
p_H2 = updated_odds_H2 / total_odds
p_H3 = updated_odds_H3 / total_odds

print(f"Probabilidades relativas: H1={p_H1:.2f}, H2={p_H2:.2f}, H3={p_H3:.2f}")
        
```

Este cuarto ejemplo ilustra la comparación y actualización de momios entre múltiples hipótesis ante nueva evidencia, mostrando cómo diferentes piezas de información pueden influir de manera variada en cada hipótesis.

```python

# Ejemplo 5: Influencia de prioridades en la actualización de momios
import numpy as np
import matplotlib.pyplot as plt

# Probabilidades previas con distintos grados de certeza
priors = [0.2, 0.5, 0.8]
odds = [p / (1 - p) for p in priors]

# Evidencia fuerte a favor de la hipótesis
p_E_given_H = 0.9
p_E_given_notH = 0.1

# Actualización de momios
updated_odds = [o * (p_E_given_H / p_E_given_notH) for o in odds]
updated_probabilities = [o / (1 + o) for o in updated_odds]

# Visualización de los resultados
plt.figure(figsize=(8, 4))
plt.plot(priors, updated_probabilities, marker='o', linestyle='-', color='b')
plt.title('Influencia de Probabilidades Previas en la Actualización de Momios')
plt.xlabel('Probabilidad Previa')
plt.ylabel('Probabilidad Posterior')
plt.grid(True)
plt.show()
    
```


Este quinto ejemplo destaca la influencia de las probabilidades previas en la actualización de los momios. Muestra cómo las creencias iniciales fuertes o débiles pueden alterar significativamente los resultados finales después de considerar la misma evidencia, ilustrando la importancia de la elección de prioridades en la inferencia Bayesiana.
